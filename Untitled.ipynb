```python
sigma = 0.1
def U_0(x):
    return np.exp(-0.5 * np.square((x - 0.5) / sigma))
```

```python
x_band = np.linspace(0, 1, 100)
plt.figure(figsize=[8, 5])
plt.title('Inintial conditions U_0(x) = U(t=0, x)')
plt.xlabel('x')
plt.ylabel('U_0')
plt.tight_layout()
plt.grid()
plt.plot(x_band, U_0(x_band))
```

## Схема уголок upwind 

```python
def triangle_downwind(a, U_0, M, K, save_period):
    x_band = np.linspace(0, 1, M + 1)
    t_band = np.linspace(0, 1, K + 1)
    h = 1. / M
    tau = 1. / K
    c = a * tau / h
    U_prev = np.array(U_0(x_band))
    output = np.array(U_0(x_band))
    
    for k in range(K):
        U_next = [U_prev[m] + c * (U_prev[m-1] - U_prev[m]) if m != 0
                  else U_prev[0] + c * (U_prev[0] - U_prev[1])
                  for m in range(M + 1)]                      
        
        if k % save_period == 0:
            U_next = np.array(U_next)
            output = np.vstack((output, U_next))
            
        U_prev = U_next.copy()
    
    return output
```

```python
M = 100
K = 10000
a = 0.3
save_period = 100
x_band = np.linspace(0, 1, M+1)
output = triangle_downwind(a, U_0, M, K, save_period)
plt.figure(figsize=[8, 5])
plt.title('U(t=1, x) from downwind triangle, a = 0.5')
plt.xlabel('x')
plt.ylabel('U')
plt.tight_layout()
plt.grid()
plt.plot(x_band, output[100, ])
print(output.shape)
```

```python
k = np.arange(0, 101)
m = np.arange(0, M + 1)
def arr_U(k, m):
    return output[k, m]
fig = plt.figure(figsize = [8, 8])
ax = plt.axes(projection='3d')
X_0 = np.linspace(0, 1, M+1)
T_0 = np.linspace(0, 1, 101)
X, T = np.meshgrid(X_0, T_0)
k, m = np.meshgrid(k, m)
data_plot = arr_U(k, m)

ax.plot_surface(T, X, data_plot, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Solution of the equation')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('U')
```

```python
def triangle_upwind(a, U_0, M, K, save_period):
    x_band = np.linspace(0, 1, M + 1)
    t_band = np.linspace(0, 1, K + 1)
    h = 1. / M
    tau = 1. / K
    c = a * tau / h
    U_prev = np.array(U_0(x_band))
    output = np.array(U_0(x_band))
    
    for k in range(K):
        U_next = [U_prev[m] + c * (U_prev[m] - U_prev[m+1]) if m != M 
                  else U_prev[M] + c * (U_prev[M-1] - U_prev[M])
                  for m in range(M + 1)]                       
        
        if k % save_period == 0:
            U_next = np.array(U_next)
            output = np.vstack((output, U_next))
            
        U_prev = U_next.copy()
    
    return output
```

```python
M = 100
K = 10000
a = 0.05
save_period = 100
x_band = np.linspace(0, 1, M+1)
output = triangle_upwind(a, U_0, M, K, save_period)
plt.figure(figsize=[8, 5])
plt.title('U(t=1, x) from upwind triangle, a = 0.05')
plt.xlabel('x')
plt.ylabel('U_0')
plt.tight_layout()
plt.grid()
plt.plot(x_band, output[100, ])
print(output.shape)
```
