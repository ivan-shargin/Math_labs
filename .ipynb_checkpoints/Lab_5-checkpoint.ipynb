```python
import numpy as np
import matplotlib.pyplot as plt
import math
from functools import reduce
from sklearn import linear_model
```

```python
def f_test(x):
    return np.sin(x) / x
```

```python
def f_easy_test(x):
    return np.sin(x)
```

```python
def left_rect(f, section, h):
    N = (section[1] - section[0]) / h
    N = int(N)
    x_grid = np.linspace(section[0], section[1], N + 1)
    
    S = 0
    for x in x_grid:
        if x != section[1]:
            S += f(x) * h 
    
    return(S)
```

```python
def grid_left_rect(f_grid, x_grid):
    S = 0
    for x in x_grid:
        if x != x_grid[-1]:
            S += f(x) * h 
    
    return(S)
```

```python
h_arr = np.linspace(0.0001, 0.2, 100)
error = []
for h in h_arr:
    I = left_rect(f_easy_test, np.array([0, math.pi]), h)
    I_precise = 2
    error.append(np.linalg.norm(I - I_precise))

error = np.array(error)
ln_error = np.log(error)
ln_h = np.log(h_arr)

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(ln_h, ln_error, s=5)
plt.xlabel("ln_h")
plt.ylabel("ln_error")
plt.title("Left rectangles, error")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(ln_h.reshape((ln_h.size, 1)), ln_error)
x0 = np.linspace(-10, -1, 100).reshape((100, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)
```

```python
def right_rect(f, section, h):
    N = (section[1] - section[0]) / h
    N = int(N)
    x_grid = np.linspace(section[0], section[1], N + 1)
    
    S = 0
    for x in x_grid:
        if x != section[0]:
            S += f(x) * h 
    
    return(S)
```

```python
def grid_right_rect(f_grid, x_grid):
    S = 0
    for x in x_grid:
        if x != x_grid[0]:
            S += f(x) * h 
    
    return(S)
```

```python
h_arr = np.linspace(0.0001, 0.2, 100)
error = []
for h in h_arr:
    I = right_rect(f_easy_test, np.array([0, math.pi]), h)
    I_precise = 2
    error.append(np.linalg.norm(I - I_precise))

error = np.array(error)
ln_error = np.log(error)
ln_h = np.log(h_arr)

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(ln_h, ln_error, s=5)
plt.xlabel("ln_h")
plt.ylabel("ln_error")
plt.title("Right rectangles, error")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(ln_h.reshape((ln_h.size, 1)), ln_error)
x0 = np.linspace(-10, -1, 100).reshape((100, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)
```
